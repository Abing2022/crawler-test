```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

req = requests.get('https://en.wikipedia.org/wiki/Los_Angeles')
webpage = req.text
soup = BeautifulSoup(webpage, 'html.parser')
paragraph = soup.find_all('p')
data = {"title":[], "href":[]}
df = pd.DataFrame(data)

for i in range(len(paragraph)):
    paragraph[i].find_all('a')
    paragraph[i].find_all('a', attrs={"title":True})
    data = {"title":[], "href":[]}
    for link in paragraph[i].find_all('a', attrs={"title":True}):
        data["title"].append(link["title"])
        data["href"].append(link["href"])
    df_new = pd.DataFrame(data)
    df = pd.concat([df, df_new], ignore_index=True, sort=False)

df.to_csv(r"C:\Users\...\Desktop\sample_data.csv")
```
